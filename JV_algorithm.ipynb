```python
def lap_jv(cost_matrix: list, dim: int) -> tuple:
    rowsol, colsol = [-1] * dim, [-1] * dim
    matches = [0] * dim   # 每一行匹配的数量
    u, v = [0] * dim, [0] * dim # u_i, v_j是对偶问题的变量, u_i + v_j <= c_ij: min c * x -> max u + v
    
    # column reduction
    for j in range(dim - 1, -1, -1):
        (vmin, imin) = min([(cost_matrix[i][j], i) for i in range(dim)])
        matches[imin] += 1
        v[j] = vmin
        
        if matches[imin] == 1:
            # assign row
            colsol[j] = imin
            rowsol[imin] = j
        elif vmin < v[rowsol[imin]]:
            colsol[j] = imin
            rowsol[imin] = j
            colsol[rowsol[imin]] = -1
        else:
            # column not assigned
            pass 
        
    # reduction transfer
    # from column to row 
    free = []
    for i in range(dim):
        if matches[i] == 0:
            free.append(i)
        elif matches[i] == 1:
            j1 = rowsol[i]
            mu = min([cost[i][j] - v[j] for j in range(dim) if j != j1])
            u[i] = mu
            v[j1] -= mu
    
    # augmenting row reduction
    # from an unassigned row to an unassigned column
    loopcnt = 2
    while (loopcnt := loopcnt - 1) + 1:
        k = 0
        prv_numfree = len(free)
        numfree = 0
        while k < prv_numfree:
            i = free[k]
            k += 1
            reduced = [cost[i][j] - v[j] for j in range(dim)]
            jmin = reduced.index(min(reduced))
            umin = reduced.pop(jmin)
            j2nd_min = reduced.index(min(reduced))
            u2nd_min = reduced.pop(j2nd_min)
            
            i0 = colsol[jmin]
            
            if u2nd_min > umin:
                # change the reduction of the minimum column to 
                # increase the minimum reduced cost in the row to the subminimum.
                v[jmin] -= u2nd_min - umin
            elif i0 > -1: 
                # j1已经有匹配边
                # swap columns j1 and j2, as j2 may be unassigned.
                jmin = j2nd_min
                i0 = colsol[jmin]
            
            # re-assign i to jmin, possibly de-assign i0
            rowsol[i] = jmin
            colsol[jmin] = i
            
            if i0 > -1:
                # put in current k, and go back to that k.
                # continue augmenting path i - j1 with i0.
                if u2nd_min > umin:
                    k -= 1
                    free[k] = i0
                else:
                    # no further augmenting reduction possible.
                    # store i0 in list of free rows for next phase.
                    free[numfree] = i0
                    numfree += 1
                    
    #
                    
            
                
        
        
            
    
        
        
```
