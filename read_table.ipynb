# Mineru

```python
json_file = "../ragflow/deepdoc/基金研报/auto/基金研报_content_list.json"
```

```python
import json 

with open(json_file, "r") as fp:
    contents = json.load(fp)
```

```python
for content in contents:
    if content['type'] == 'table':
        body = content['table_body']
        tbl = pd.read_html(body)[0]
        print(f"""
            On page {content['page_idx']}, 
            titled {content['table_caption'][0]}, 
            with foodnote {content['table_footnote'][0]}
        """)
        print(tbl, end='\n\n')
```
