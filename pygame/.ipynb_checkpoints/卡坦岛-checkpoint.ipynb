```python
import pygame
import random
import math
import numpy as np 

# 常量，屏幕宽高
WIDTH, HEIGHT = 800, 600

# 初始化操作
pygame.init()
pygame.mixer.init()
# 创建游戏窗口
screen = pygame.display.set_mode((WIDTH, HEIGHT))

# 设置游戏标题
pygame.display.set_caption('妹多局之卡坦岛')
```

```python
red, blue, yellow = [pygame.Color(color) for color in ['red', 'blue', 'yellow']]
```

```python
help(pygame.Surface)
```

```python
# 添加系统时钟，用于设置帧的刷新
FPS = 10
clock = pygame.time.Clock()

# 定义class
class Vertice(pygame.sprite.Sprite):
    WIDTH, HEIGHT = 10, 10
    def __init__(self, x, y):
        super().__init__()
        self.rect = pygame.rect.Rect(x, y, self.WIDTH, self.HEIGHT)
        self.image = pygame.Surface(self.WIDTH, self.HEIGHT)
        
        # 用颜色填充
        colorImage = pygame.Surface(self.image.get_size()).convert_alpha()
        colorImage.fill(red)
        self.image.blit(colorImage, (x, y), special_flags = pygame.BLEND_RGBA_MULT)


        
class Hexagon(pygame.sprite.Sprite):
    RADIUS = 40 
    RESOURCE_TYPES = 'sheep wood wheat mineral brick'.split(' ')
    RESOURCE_IMAGES = [pygame.image.load(f"./images/catan/{res}.jpeg") for res in RESOURCE_TYPES]
    TIPO2IMAGE = dict(zip(RESOURCE_TYPES, RESOURCE_IMAGES))
    WIDTH, HEIGHT = RADIUS * math.sqrt(3), RADIUS
    
    def __init__(self, resource_type, topleft_x, topleft_y):
        super().__init__()
        self.res = resource_type
        self.image = self.TIPO2IMAGE[self.res]
        self.image = pygame.transform.scale(self.image, (self.WIDTH, self.HEIGHT))
        self.rect = pygame.rect.Rect(topleft_x, topleft_y, self.WIDTH, self.HEIGHT)
```

```python
# 初始化精灵组
hex_group = pygame.sprite.Group()
ver_group = pygame.sprite.Group()

# 定义人物
hexs = [
    Hexagon(
            topleft_x=math.sqrt(3) * Hexagon.RADIUS * (i+1),
            topleft_y=Hexagon.RADIUS,
            resource_type=random.choice(Hexagon.RESOURCE_TYPES)
    ) for i in range(6)
]

for ele in hexs:
    hex_group.add(ele)
```

```python
# 保持游戏运行状态(游戏循环）
while True:
    # ===========游戏帧的刷新===========
    clock.tick(FPS)

    # 检测事件
    for event in pygame.event.get():
        # 检测关闭按钮被点击的事件
        if event.type == pygame.QUIT:
            # 退出
            pygame.quit()
            exit()
    
    screen.fill((0,0,0))
    hex_group.draw(screen)
    
    pygame.display.update()
```
